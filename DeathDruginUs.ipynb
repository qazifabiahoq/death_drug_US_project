```python
# Import dependencies
from pymongo import MongoClient
from pprint import pprint
from decimal import Decimal

# Create an instance of MongoClient
mongo = MongoClient(port=27017)

# confirm that our new database was created
database_names = mongo.list_database_names()
print("Databases:", database_names)
```

```python
# Assign the "death_data" and "drug_data" databases to variables
death_db = mongo["death_data"]
drug_db = mongo["drug_data"]
```

```python
# Review the collections in the "death_data" database
death_collections = death_db.list_collection_names()
print("Collections in 'death_data' database:", death_collections)

# Review the collections in the "drug_data" database
drug_collections = drug_db.list_collection_names()
print("Collections in 'drug_data' database:", drug_collections)
```

```python
# Access the "death_data" database
death_db = mongo['death_data']

# Access the "drug_overdose_death_rates" collection
drug_overdose_collection = death_db['drug_overdose_death_rates']

# Print a document from the "drug_overdose_death_rates" collection
pprint(drug_overdose_collection.find_one())

```

```python
# Access the "suicide_death_rates" collection
suicide_death_collection = death_db['suicide_death_rates']

# Print a document from the "suicide_death_rates" collection
pprint(suicide_death_collection.find_one())
```

```python
# Access the "drug_overdose_death_rates" collection
drug_overdose_death_collection = drug_db['drug_overdose_death_rates']

# Print a document from the "drug_overdose_death_rates" collection
pprint(drug_overdose_death_collection.find_one())
```

```python
# Limit the number of documents returned by the query to 100
for document in drug_overdose_death_collection.find().limit(100):
    pprint(document)
```

```python
# This is the unique years present in the suicide_death_rates collection
unique_years_suicide = death_db.suicide_death_rates.distinct("YEAR")
print("Unique years in suicide_death_rates:", unique_years_suicide)
```

```python
# This is years present in the drug_overdose_death_rates collection
unique_years_drug_overdose = drug_db.drug_overdose_death_rates.distinct("YEAR")
print("Unique years in drug_overdose_death_rates:", unique_years_drug_overdose)
```
